```python
import pandas as pd
from bs4 import BeautifulSoup

soup = BeautifulSoup(open("/Users/rhey.magcalas/Downloads/Airtable - Public Capstone Projects.html"), "html.parser")

result = {}
result['Fellow Name'] = []
result['Capstone Title'] = []
result['Cohort'] = []
result['Topic 1'] = []
result['Topic 2'] = []
result['Topic 3'] = []

for div in soup.findAll('div', attrs={'class':'baymaxGalleryCard'}):
    topic_1 = ''
    topic_2 = ''
    topic_3 = ''
    fellow_name = div.find('a').contents[0]
    capstone_title = div.findAll("div", {"class": "truncate"})[1].contents[0]
    cohort = div.findAll("div", {"class": "truncate"})[len(div.findAll("div", {"class": "truncate"}))-1].contents[0]
    
    result['Capstone Title'].append(capstone_title)
    result['Fellow Name'].append(fellow_name)
    result['Cohort'].append(cohort)
    
    current_counter = 0
    counter = 4
    while len(div.findAll("div", {"class": "truncate"}))-2 > counter:
        current_counter += 1
        if current_counter == 1:
            topic_1 = div.findAll("div", {"class": "truncate"})[counter].contents[0]
        elif current_counter == 2:
            topic_2 = div.findAll("div", {"class": "truncate"})[counter].contents[0]
        else:
            topic_3 = div.findAll("div", {"class": "truncate"})[counter].contents[0]
        counter +=1
    result['Topic 1'].append(topic_1)
    result['Topic 2'].append(topic_2)
    result['Topic 3'].append(topic_3)
    
df = pd.DataFrame.from_dict(result)
df[df['Cohort']=='Cohort 5'].head(16)
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('capstones.db')

conn.execute('''CREATE TABLE COHORT
         (ID INTEGER PRIMARY KEY AUTOINCREMENT  NOT NULL ,
         Cohort         TEXT    NOT NULL;''')

conn.execute('''CREATE TABLE FELLOWS
         (ID INTEGER PRIMARY KEY AUTOINCREMENT  NOT NULL ,
         FIRST_NAME  CHAR(50)  NOT NULL,
         LAST_NAME  CHAR(50)  NOT NULL;''')

conn.execute('''CREATE TABLE CAPSTONES
         (ID INTEGER PRIMARY KEY AUTOINCREMENT  NOT NULL ,
         TITLE  CHAR(250)  NOT NULL''')

conn.execute('''CREATE TABLE TOPICS
         (ID INTEGER PRIMARY KEY AUTOINCREMENT  NOT NULL ,
         TOPICS  CHAR(250)  NOT NULL''')


conn.execute('''CREATE TABLE CAPSTONE_TOPICS
         (ID INTEGER PRIMARY KEY AUTOINCREMENT  NOT NULL ,
         FOREIGN KEY (CAPSTONESID) REFERENCES CAPSTONES(ID),
         FOREIGN KEY (TOPICSID) REFERENCES TOPICS(ID)''')


conn.close()
```
